# Matplotlib Part 1

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
%matplotlib inline
```

```python
x = np.linspace(0,5,11)
y = x**2
```

```python
x
```

```python
y
```

```python
fig = plt.figure()

axes1 = fig.add_axes([0.1,0.1,0.7,.7])
axes2 = fig.add_axes([0.2,0.4,.25,0.25])

axes1.plot(x,y)
axes1.set_title("Bigger Plot")

axes2.plot(y,x)
axes2.set_title("Smaller Plot")
```

# Matplotlib Part 2

```python
fig,axes = plt.subplots(nrows=1, ncols=2)

axes[0].plot(x,y)
axes[0].set_title('First Plot')

axes[1].plot(y,x)
axes[1].set_title('Second Plot')
```

## Figure size and DPI

```python
fig,axes = plt.subplots(nrows=2, ncols=1, figsize=(8,2))

axes[0].plot(x,y)

axes[1].plot(y,x)

plt.tight_layout()
```

```python
fig.savefig('playgroundfig.png', dpi=200)
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])
ax.plot(x,x**2,label='X*2')
ax.plot(x,x**3,label='X*3')

ax.legend(loc=0)
```

# Matplotlib Part 3

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,y,color="#6AB5A6",lw=3,ls='--')

ax.set_xlim([0,1])
ax.set_ylim([0,1.5])
```
