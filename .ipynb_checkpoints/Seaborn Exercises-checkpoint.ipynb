# Seaborn Exercises

### The Data

```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
titanic = sns.load_dataset('titanic')
```

```python
titanic.head()
```

## Exercises

```python
sns.set(font_scale=1.1)
sns.set_style('whitegrid')
grid = sns.JointGrid(data=titanic,x='fare',y='age',height=6)
grid.plot(sns.scatterplot,sns.histplot)
```

```python
sns.histplot(data=titanic,x='fare',bins=30)
```

```python
sns.boxplot(data=titanic,x='class',y='age')
```

```python
sns.swarmplot(data=titanic,x='class',y='age')
```

```python
sns.countplot(data=titanic,x='sex')
```

```python
titanic.corr()
```

```python
sns.heatmap(data=titanic.corr(),cmap='coolwarm')
```

```python
grid = sns.FacetGrid(titanic,col='sex')
grid.map(sns.histplot,'age',bins=10)
```
