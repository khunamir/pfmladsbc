### Bank Of America

```python
start = datetime.datetime(2006,1,1)
end = datetime.datetime(2016,1,1)
BAC = data.DataReader("BAC", 'yahoo', start=start, end=end)
BAC.head()
```

### CitiGroup

```python
start = datetime.datetime(2006,1,1)
end = datetime.datetime(2016,1,1)
Citi = data.DataReader("C", 'yahoo', start=start, end=end)
Citi.head()
```

### Goldman Sachs

```python
start = datetime.datetime(2006,1,1)
end = datetime.datetime(2016,1,1)
GS = data.DataReader("GS", 'yahoo', start=start, end=end)
GS.head()
```

### JPMorgan Chase

```python
start = datetime.datetime(2006,1,1)
end = datetime.datetime(2016,1,1)
JPM = data.DataReader("JPM", 'yahoo', start=start, end=end)
JPM.head()
```

### Morgan Stanley

```python
start = datetime.datetime(2006,1,1)
end = datetime.datetime(2016,1,1)
MS = data.DataReader("MS", 'yahoo', start=start, end=end)
MS.head()
```

### Wells Fargo

```python
start = datetime.datetime(2006,1,1)
end = datetime.datetime(2016,1,1)
WFC = data.DataReader("WFC", 'yahoo', start=start, end=end)
WFC.head()
```

#### Create a list of banks' tickers

```python
tickers = ['BAC','C','GS','JPM','MS','WFC']
```

#### Use pd.concat to concat the dataframes together. Set the key argument to equal the tickers list

```python
bank_stocks = pd.concat([BAC,Citi,GS,JPM,MS,WFC], axis=1, keys=tickers)
```

#### Set the column name for each level

```python
bank_stocks.columns.names = ['Bank Ticker','Stock Info']
```

#### Check the head of the dataframe

```python
bank_stocks.head()
```

## Exploratory Data Analysis (EDA)

#### What is the max close price for each bank stocks for the time period?

```python
bank_stocks.xs(key='Close',axis=1,level='Stock Info').max()
```

** Create a new empty DataFrame called returns. This dataframe will contain the returns for each bank's stock. returns are typically defined by:**

$$r_t = \frac{p_t - p_{t-1}}{p_{t-1}} = \frac{p_t}{p_{t-1}} - 1$$

```python
returns = pd.DataFrame()
```

** We can use pandas pct_change() method on the Close column to create a column representing this return value. Create a for loop that goes and for each Bank Stock Ticker creates this returns column and set it as a column in the returns DataFrame.**

```python
for tick in tickers:
    returns[tick + " Return"] = bank_stocks[tick]['Close'].pct_change()
    
returns.head()
```
